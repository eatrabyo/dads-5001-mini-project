```python
import pandas as pd 
import numpy as np

us_arrival = pd.read_excel(r"data_source\us-Annual-Arrivals-2000-to-Present–Country-of-Residence.xlsx")
df = us_arrival
dfm = df.melt(id_vars=['Countries','World Region','Comment','Notes:'])
dfm = dfm.rename(columns={'variable':'year','value':'count'})
dfmr = dfm.loc[dfm['count'].notna() & dfm['World Region'].isna(),:]
dfmr = dfmr.sort_values(['Countries','year'], ascending=[True,True])
count_byregion = dfmr.loc[:,['Countries','year','count']]
count_byregion
```

```python
dfmc = dfm[dfm['World Region'].notna() & dfm['count'].notna()]
dfmc = dfmc.drop(['Notes:','Comment'],axis=1)
dfmc
```

```python
dfm_china = dfmc.loc[dfmc['Countries']=='China',:].reset_index(drop=True)
dfm_china
```
